```python
import pandas as pd
from gnomad_db.database import gnomAD_DB
```

```python
# pass dir
database_location = "test_dir"
db = gnomAD_DB(database_location, gnomad_version="v3")
```

```python
# get some variants
var_df = pd.read_csv("../output/VCFs/sample_False_indels.vcf", sep="\t", names=db.columns, index_col=False, comment = "#")
```

```python
var_df = "../output/VCFs/sample_False_indels.vcf"

# Initialize an empty list to store data lines
data_lines = []

# Read the VCF file line by line, skipping lines where CHROM starts with '#'
with open(var_df, 'r') as f:
    for line in f:
        if line.startswith('#'):
            continue  # Skip header lines
        if line.split('\t')[0].startswith('#'):
            continue  # Skip lines where CHROM starts with '#'
        data_lines.append(line.strip())

variants_df = pd.read_csv(
    pd.Series(data_lines, name=db.columns, index_col = False),
    sep='\t',
    comment='#',  # Treat '#' as a comment to ignore any remaining header lines
    header=None   # No header because we're reading just the data lines
)
```

```python
# IMPORTANT: The database removes internally chr prefix (chr1->1)
# insert these variants
db.insert_variants(var_df)
```
